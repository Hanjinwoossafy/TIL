```python
#### 월
# 입력 받았다 치고
for tc in range(1,11):  # 테스트 케이스가 10개라고 했음
    # 모든 빌딩에 대해서 조망권이 있는 세대수 계산해서 더하기
    N = int(input())
    buildings = list(map(int,input().split()))
    result = 0    #누적합을 구해야 하기 때문에... 초기값은 0
    for i in range(2,N-2): # i : 빌딩번호
        # i번 건물의 높이 : A
        A = buildings[i]
        # 주변 네개 건물 중 가장 높은 건물 높이 B
        neighbors = [buildings[i-2],buildings[i-1],buildings[i+1],buildings[i+2]]
        max_v = neighbors[0]
        for j in range(4):
            if max_v < neighbors[j]:
                max_v = neighbors[j]
        B = max_v
        # result += A - B if A > B else 0
        if A > B:   # 조망권이 있는 경우에만 더하기
            result += A - B

    print(f'#{tc} {result}')
```

```python
####화
# 카운팅 정렬  => 요소가 나온 횟수를 세어서 위치를 알아내어 정렬하는 방법
arr = [4,5,3,1,6,4,2,6,6,2]
N = 10
K = 6
# 1. arr 의 각 요소의 개수세기
# 2. 개수의 누적합 구하기 (요소의 위치를 구하기 위해서)
# 3. 각 요소 위치에 맞게 복사해주기
counts = [0] * (K+1)
# 크기 몇 짜리로 해야 합니까? 개수를 세는 변수니까 초기화는 0으로
# 1. arr 의 각 요소의 개수세기
for i in range(N):
    #  arr[i] 얘가 몇개 나왔는지 세야 합니다.
    counts[arr[i]] += 1
# 2. 개수의 누적합 구하기 (요소의 위치를 구하기 위해서)
# [0, 1, 2, 1, 2, 1, 3]
print(counts)
for i in range(1,K+1):
    counts[i] = counts[i] + counts[i-1]
    # counts[i] += counts[i-1]
print(counts)
# 3. 각 요소 위치에 맞게 복사해주기
sorted_arr = [0] * N
for i in range(N-1,-1,-1):
    # arr[i] 요소가 어디 들어가는지 찾아서 넣어주기
    counts[arr[i]] -= 1 #실제 인덱스를 찾기위해서 -1
    # counts[arr[i]] 가 arr[i] 가 들어갈 인덱스
    sorted_arr[counts[arr[i]]] = arr[i]

print(sorted_arr)
```

```python
# 1. 주어진 배열에서 최대값과 최소값 의 위치를 구함
T = 10
for tc in range(1,T+1):
    N = int(input())
    boxes = list(map(int,input().split()))

    for _ in range(N):
        # 2. 최대값 위치, 최소값위치 찾아서
        # 최대값 위치 -1 ,최소값 위치 +1 작업을 N번 반복
        max_idx = 0
        min_idx = 0
        for i in range(100):
            if boxes[i] > boxes[max_idx]:
                max_idx = i
            if boxes[i] < boxes[min_idx]:
                min_idx = i
        boxes[max_idx] -= 1
        boxes[min_idx] += 1
    # N 번의 작업이 끝남
    # 3. 마지막 최대 최소 위치 구해서 정답 출력하기
    max_idx = 0
    min_idx = 0
    for i in range(100):
        if boxes[i] > boxes[max_idx]:
            max_idx = i
        if boxes[i] < boxes[min_idx]:
            min_idx = i
    print(f'#{tc} { boxes[max_idx] - boxes[min_idx]}')
```

```python
def find_min_max(boxes):
    max_idx = 0
    min_idx = 0
    for i in range(100):
        if boxes[i] > boxes[max_idx]:
            max_idx = i
        if boxes[i] < boxes[min_idx]:
            min_idx = i
    return max_idx,min_idx

# 1. 주어진 배열에서 최대값과 최소값 의 위치를 구함
T = 10
for tc in range(1,T+1):
    N = int(input())
    boxes = list(map(int,input().split()))

    for _ in range(N):
        # 2. 최대값 위치, 최소값위치 찾아서
        # 최대값 위치 -1 ,최소값 위치 +1 작업을 N번 반복
        max_idx, min_idx = find_min_max(boxes)
        boxes[max_idx] -= 1
        boxes[min_idx] += 1
    # N 번의 작업이 끝남
    # 3. 마지막 최대 최소 위치 구해서 정답 출력하기
    max_idx, min_idx = find_min_max(boxes)
    print(f'#{tc} { boxes[max_idx] - boxes[min_idx]}')
```
